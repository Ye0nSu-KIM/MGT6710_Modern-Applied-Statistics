# 2. Ridge penalty

### (1) Use sl & sw as explanatory variables 

- standardization

```python
x_var = iris[['sl', 'sw']]

x_var = (x_var - x_var.mean())/x_var.std()
```

```python
R_Xvar_1 = sm.add_constant(x_var)
```

```python
R_model_1 = sm.Logit(yvar, R_Xvar_1)
```

#### alpha cross validation : Find alpha which makes the smallest AIC

```python
alpha_cv = np.linspace(0,100,200)
def finding_alpha_using_aic(Ridge_logit, alpha_cv):
    AIC = []
    for i in alpha_cv:
        results = Ridge_logit.fit_regularized(start_params=None,alpha=i,L1_wt =0) # L1_wt = 0 이면 Ridge
        AIC.append(results.aic)
    return alpha_cv[np.argmin(AIC)]
```

```python
alpha = finding_alpha_using_aic(R_model_1, alpha_cv)
```

```python
R_res_1 = R_model_1.fit_regularized(start_paras=None,alpha = alpha, L1_wt =0)
```

```python
R_res_1.summary()
```

```python
R_predict_1 = np.where(R_res_1.predict(R_Xvar_1)>=0.5,1,0)
```

```python
R_predict_1
```

```python
print(confusion_matrix(yvar,R_predict_1))
```
